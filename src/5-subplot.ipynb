### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

# Create a figure and 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Data for plots
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]  # Squared values for line plot

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
bar_colors = ['red', 'blue', 'green', 'purple']

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
pie_colors = ['blue', 'red', 'green', 'purple']

# Line Plot (Top-Left)
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='blue')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y values")
axs[0, 0].grid(True)

# Bar Chart (Top-Right)
axs[0, 1].bar(categories, values, color=bar_colors)
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter Plot (Bottom-Left)
axs[1, 0].scatter(scatter_x, scatter_y, color='blue', marker='o')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X values")
axs[1, 0].set_ylabel("Y values")
axs[1, 0].grid(True)

# Pie Chart (Bottom-Right)
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=pie_colors, startangle=140, wedgeprops={'edgecolor': 'black'})
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
